## Check GPU things

```python
# Checks platform and reports what hardware backends are available

class color:
   PURPLE = '\033[1;35;48m'
   CYAN = '\033[1;36;48m'
   BOLD = '\033[1;37;48m'
   BLUE = '\033[1;34;48m'
   GREEN = '\033[1;32;48m'
   YELLOW = '\033[1;33;48m'
   RED = '\033[1;31;48m'
   BLACK = '\033[1;30;48m'
   UNDERLINE = '\033[4;37;48m'
   END = '\033[1;37;0m'


try:
    import torch
except ImportError:
    pytorch_installed = False
    cformat = color.RED
else:
    pytorch_installed = True
    
if pytorch_installed:
    if torch.backends.mps.is_available():
        mps_device = torch.device("mps")
        x = torch.ones(1, device=mps_device)
        pytorch_gpu = True
        cformat = color.GREEN
    else:
        pytorch_gpu = False
        cformat = color.YELLOW

print( "Platform Descipription:")
print( "Name    |  CPU  |  GPU")
print( "----------------------")

print(cformat + "pytorch " + color.END + "|", end='')

if pytorch_installed:
    if pytorch_gpu:
        print(   "       |  ", u'\u2713')
    else:
        print(   "  ", u'\u2713', "  |    ")
else:
    print(       "       |")
        
        
try:
    from jax.lib import xla_bridge
except ImportError:
    jax_installed = False
    cformat = color.RED
else:
    jax_installed = True
    
if jax_installed:
    platform = xla_bridge.get_backend().platform
    if platform == 'cpu':
        jax_gpu = False
        cformat = color.YELLOW
    else:
        jax_gpu = True
        cformat = color.GREEN
        
print(cformat + "jax     " + color.END + "|", end='')

if jax_installed:
    if jax_gpu:
        print(   "  ", u'\u2713', "  |  ", u'\u2713')
    else:
        print(   "   X   |    ")
else:
    print(       "       |")
```
